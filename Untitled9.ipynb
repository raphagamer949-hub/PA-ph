```python
# questao 4

# Um programa que verifica se a pessoa é criança, adolescente, Adulto ou Idoso

# Entrada
idade = int(input("Digite sua idade: "))

# Processamento e Saída
if idade >= 0 and idade <= 12:
  print(f"Com {idade} anos, você é uma criança.")
elif idade >= 13 and idade <= 17:
  print(f"Com {idade} anos, você é um adolescente.")
elif idade >= 18 and idade <= 59:
  print(f"Com {idade} anos, você é um adulto.")
elif idade >= 60:
  print(f"Com {idade} anos, você é um idoso.")
else:
  print("Idade inválida. Por favor, digite uma idade não negativa.")
```

```python
# questao 6

# Programa para verificar se um número é positivo, negativo ou zero

# Entrada
numero = float(input("Digite um número: "))

# Processamento e Saída
if numero > 0:
    print(f"O número {numero} é positivo.")
elif numero < 0:
    print(f"O número {numero} é negativo.")
else:
    print(f"O número {numero} é zero.")
```

```python
# questao 7

# @title Texto de título padrão
# Operadores aritiméticos

#Fazer um programa que 02 numeros
#Saida mostrar operáções aritimeticas

# Entrada

a = int(input("digite um numero: "))
b = int(input("digite um numero: "))

#Processamento

SOMA = a + b
SUBTRACAO = a - b
MULTIPLICACAO = a * b
DIVISAO = a / b
divisao_inteira = a // b
RESTO = a % b
POTENCIA = a ** b

#Saida
print("")
print("               NUMEROS ARITIMETICOS           ")
print("")
print("                      SOMA                    ")
print("")
print(f"o resultado de {a} mais {b} é igual a: {SOMA}")
print("----------------------------------------------")
print("                   SUBTRACAO                  ")
print("")
print(f"o resultado de {a} menos {b} é igual a: {SUBTRACAO}")
print("----------------------------------------------")
print("                  MULTIPLCACAO                ")
print("")
print(f"o resultado de {a} vezes {b} é igual a: {MULTIPLICACAO}")
print("----------------------------------------------")
print("                      DIVISAO                  ")
print("")
print(f"o resultado de {a} dividido por {b} é igual a: {DIVISAO}")
print("----------------------------------------------")
print("                      DIVISAO INTEIRA                    ")
print("")
print(f"o resultado de {a} dividido por {b} é igual a: {divisao_inteira}")
print("----------------------------------------------")
print("                      RESTO                   ")
print("")
print(f"o resultado de {a} mais {b} é igual a: {RESTO}")
print("----------------------------------------------")
print("                      POTENCIA                   ")
print("")
print(f"o resultado da potencia de {a} e {b} é {POTENCIA}")
print("----------------------------------------------")
```

```python
# questao 8

# Programa para mostrar o dia da semana correspondente a um número

# Entrada
numero_dia = int(input("Digite um número de 1 a 7: "))

# Processamento e Saída
if numero_dia == 1:
  print("O dia correspondente é Domingo.")
elif numero_dia == 2:
  print("O dia correspondente é Segunda-feira.")
elif numero_dia == 3:
  print("O dia correspondente é Terça-feira.")
elif numero_dia == 4:
  print("O dia correspondente é Quarta-feira.")
elif numero_dia == 5:
  print("O dia correspondente é Quinta-feira.")
elif numero_dia == 6:
  print("O dia correspondente é Sexta-feira.")
elif numero_dia == 7:
  print("O dia correspondente é Sábado.")
else:
  print("Número inválido. Por favor, digite um número entre 1 e 7.")
```

```python
# questao 9

# Programa para verificar se um ano é bissexto

# Entrada
ano = int(input("Digite um ano: "))

# Processamento e Saída
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"O ano {ano} é bissexto.")
else:
  print(f"O ano {ano} não é bissexto.")
```

```python
# questao 10

# Programa para calcular desconto em compras

# Entrada
valor_compra = float(input("Digite o valor da compra: "))

# Processamento
if valor_compra >= 100:
  desconto = valor_compra * 0.10
  valor_final = valor_compra - desconto

  print(f"Valor da compra: R${valor_compra:.2f}")
  print(f"Desconto aplicado (10%): R${desconto:.2f}")
  print(f"Valor final com desconto: R${valor_final:.2f}")
elif valor_compra >= 50 and valor_compra < 100:
  desconto = valor_compra * 0.05

  valor_final = valor_compra - desconto
  print(f"Valor da compra: R${valor_compra:.2f}")
  print(f"Desconto aplicado (5%): R${desconto:.2f}")
  print(f"Valor final com desconto: R${valor_final:.2f}")
else:

  print(f"Valor da compra: R${valor_compra:.2f}")
  print("Sem desconto aplicado.")
  print(f"Valor final: R${valor_compra:.2f}")
```
