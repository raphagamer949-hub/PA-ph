```python
# @title Texto de título padrão
# Operadores aritiméticos

#Fazer um programa que 02 numeros
#Saida mostrar operáções aritimeticas

# Entrada

a = int(input("digite um numero: "))
b = int(input("digite um numero: "))

#Processamento

SOMA = a + b
SUBTRACAO = a - b
MULTIPLICACAO = a * b
DIVISAO = a / b
divisao_inteira = a // b
RESTO = a % b
POTENCIA = a ** b

#Saida
print("")
print("               NUMEROS ARITIMETICOS           ")
print("")
print("                      SOMA                    ")
print("")
print(f"o resultado de {a} mais {b} é igual a: {SOMA}")
print("----------------------------------------------")
print("                   SUBTRACAO                  ")
print("")
print(f"o resultado de {a} menos {b} é igual a: {SUBTRACAO}")
print("----------------------------------------------")
print("                  MULTIPLCACAO                ")
print("")
print(f"o resultado de {a} vezes {b} é igual a: {MULTIPLICACAO}")
print("----------------------------------------------")
print("                      DIVISAO                  ")
print("")
print(f"o resultado de {a} dividido por {b} é igual a: {DIVISAO}")
print("----------------------------------------------")
print("                      DIVISAO INTEIRA                    ")
print("")
print(f"o resultado de {a} dividido por {b} é igual a: {divisao_inteira}")
print("----------------------------------------------")
print("                      RESTO                   ")
print("")
print(f"o resultado de {a} mais {b} é igual a: {RESTO}")
print("----------------------------------------------")
print("                      POTENCIA                   ")
print("")
print(f"o resultado da potencia de {a} e {b} é {POTENCIA}")
print("----------------------------------------------")
```
